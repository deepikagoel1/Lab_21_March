```python
import pandas as pd
from matplotlib import pyplot as plt
from pandas.plotting import lag_plot
import numpy as np
from pandas.plotting import autocorrelation_plot
from statsmodels.graphics.tsaplots import plot_acf
from sklearn.metrics import mean_squared_error
import math as m
from statsmodels.tsa.ar_model import AutoReg as AR
import matplotlib.dates as mdates
from datetime import datetime
import math
```

```python
series = pd.read_csv('dataset1_HP.csv')
fig, ax = plt.subplots()
#plt.plot_date(series.index.values, series[1].values,linestyle='-')
plt.plot(series.index.values, series['HP'].values,'b-')
plt.xlabel("Index_of_Date")
plt.ylabel("mega-units")
plt.title("Lineplot")
```

```python
################ Read Data with dates #####################
series = pd.read_csv('dataset1_HP.csv', parse_dates=['Date'],index_col=['Date'],sep=',')
```

```python
################# Autocorrelation plot for limited time lag ###############
plot_acf(series, lags=10)
plt.show()
```

```python
#################### Autoregression Model ##############
X = series.values
tst_sz = m.ceil(len(X)*0.25)
train, test = X[:len(X)-tst_sz], X[len(X)-tst_sz:] # No random spilt
print("Number of days for training:",len(train),"   For testing:",len(test))

min_RMSE_AR = 1000  # A random lagre number
lag =0 # Initialize to dummy

for i in (1,5,10,15,20,50,125,150):
    ARmodel = AR(train, lags=i)
    ARmodel_fit = ARmodel.fit()
    #print('Coefficients: %s',ARmodel_fit.params)
    # make predictions
    predictions = ARmodel_fit.predict(start=len(train), end=len(train)+len(test)-1)
    error = m.sqrt(mean_squared_error(test, predictions))
    
    if error < min_RMSE_AR:
        min_RMSE_AR = error
        lag = i

print("Minimum RMSE using AR:",min_RMSE_AR)
print('Corrsponding lag:',i)
```

```python
print('Test RMSE: %.3f' % error)

plt.plot(test, color='blue')
plt.plot(predictions, color='red')
plt.xlabel("test_days")
plt.ylabel("Consumption in mega-units")
plt.title("Autoreg_predicted_vs_GT_lag_5")
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train, order=(25,1,1))
model_fit = model.fit()

predictions = model_fit.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)
error = m.sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % error)

plt.plot(test, color='blue')
plt.plot(predictions, color='red')
plt.xlabel("test_days")
plt.ylabel("Consumption in mega-units")
plt.title("Autoreg_predicted_vs_GT_lag_5")
```
